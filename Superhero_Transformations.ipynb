```python
import pandas as pd
import os, json
```

```python
info_df = pd.read_csv(r"C:\Users\Antho\Downloads\superhero_info - superhero_info.csv")
powers_df = pd.read_csv(r"C:\Users\Antho\Downloads\superhero_powers - superhero_powers.csv")
```

```python
info_df.head()
```

# Clean files and combine

```python
#Seperate hero and publisher
info_df[["Hero", "Publisher"]] = info_df["Hero|Publisher"].str.split("|", expand = True)
info_df.head()
```

```python
info_df.drop(columns = ["Hero|Publisher"], inplace = True)
info_df.head()
```

```python
#Seperate height and weight
info_df["Measurements"] = info_df["Measurements"].str.replace("'", '"')
info_df["Measurements"] = info_df["Measurements"].apply(json.loads)
info_df["Measurements"].head()
```

```python
test_measurement = info_df.loc[4, "Measurements"]
print(type(test_measurement))
```

```python
height_weight = info_df["Measurements"].apply(pd.Series)
height_weight
```

```python
info_df = pd.concat((info_df, height_weight), axis = 1)
info_df.drop(columns = ["Measurements"], inplace = True)
info_df.head()
```

```python
info_df["Height"].str.split(" ", expand = True)
info_df[["Height_cm", "cm"]] = info_df["Height"].str.split(" ", expand = True)
info_df["Weight"].str.split(" ", expand = True)
info_df[["Weight_kg", "kg"]] = info_df["Weight"].str.split(" ", expand = True)
info_df.drop(columns = ["Height", "Weight", "cm", "kg"], inplace = True)
info_df.head()
```

```python
#Change datatype to float
info_df["Height_cm"] = info_df["Height_cm"].astype("float")
info_df["Weight_kg"] = info_df["Weight_kg"].astype("float")
info_df.info()
```

```python
powers_df.head()
```

```python
powers_df["Powers_List"] = powers_df["Powers"].str.split(",")
powers_df.head()
```

```python
exploded_list = powers_df.explode("Powers_List")
exploded_list.head()
```

```python
powers = exploded_list["Powers_List"].unique()
powers
```

```python
#Make new column for every power
for i in powers:
    powers_df[i] = powers_df["Powers"].str.contains(i)
powers_df.head()
```

```python
combined_df = pd.merge(info_df, powers_df, right_on = "hero_names", left_on = "Hero")
combined_df.head()
```

# Part 2

```python
super_speed = combined_df["Super Speed"] == True
speed_df = combined_df[super_speed]
non_speed_df = combined_df[~super_speed]
speed_weight = speed_df["Weight_kg"].mean()
non_speed_weight = non_speed_df["Weight_kg"].mean()
print(f"Average weight for super hero with super speed: {speed_weight:.2f}")
print(f"Average weight for super hero without super speed: {non_speed_weight:.2f}")
```
